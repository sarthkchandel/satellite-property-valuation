```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

df = pd.read_excel("../data/raw/train.xlsx")
df = df.dropna()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["price"], bins=50, kde=True)
plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["sqft_living"], y=df["price"], alpha=0.4)
plt.title("Price vs Living Area (sqft)")
plt.xlabel("Sqft Living")
plt.ylabel("Price")
plt.show()
```

```python
img_paths = [
    "../data/images/train/house_0.png",
    "../data/images/train/house_1.png",
    "../data/images/train/house_2.png"
]

plt.figure(figsize=(10,4))
for i, path in enumerate(img_paths):
    img = Image.open(path)
    plt.subplot(1, 3, i+1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"Sample Image {i+1}")

plt.show()
```

```python
models = ["Tabular Only", "Multimodal"]
rmse_values = [126000, 120000]

plt.figure(figsize=(6,4))
sns.barplot(x=models, y=rmse_values)
plt.title("RMSE Comparison Between Models")
plt.ylabel("RMSE")
plt.show()
```
